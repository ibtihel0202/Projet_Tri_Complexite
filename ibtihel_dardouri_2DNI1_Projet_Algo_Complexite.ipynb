## 1) Algorithme par selection:

*Implémentation du tri par selection*

```python
from random import *
from time import *
def liste_alea(n):
    s=[0]*n
    for i in range(n):
        s[i]=randrange(1,100)
    return s
def tri_selection(liste):
    for i in range(0,len(liste)-1):
      # Trouver le min
        min = liste[i]
        indice=i
        for j in range(i+1, len(liste)):
            if liste[j] < min:
                min = liste[j]
                indice=j
                
        tmp = liste[i]
        liste[i] = liste[indice]
        liste[indice] = tmp

```

*Programme principale pour tester le code ci-dessus et afficher le temps d'éxécution*

```python
tab1 = liste_alea(50)
t1=time()
print ("La liste initiale est: \n",tab1)
tri_selection(tab1)
print ("La liste triée est: \n",tab1)
t2=time()
a=t2-t1
print("le temps d'éxécution de tri par selection est: \n",a)
```

## 2) Tri à bulle :

*Implémentation du Tri à bulle*

```python
from random import *
from time import *
def liste_alea(n):
    s=[0]*n
    for i in range(n):
        s[i]=randrange(1,100)
    return s
def tri_bulle(liste):
    # Traverser tous les éléments du tableau
    for i in range(len(liste)):
        for j in range(0, len(liste)-i-1):
            # échanger si l'élément trouvé est plus grand que le suivant
            if liste[j] > liste[j+1] :
                liste[j], liste[j+1] = liste[j+1], liste[j]

```

*Programme principale pour tester le code ci-dessus et afficher le temps d'éxécution*

```python
tab2 = liste_alea(50)
t1=time()
print ("La liste initiale est: \n",tab2)
tri_bulle(tab2)
print ("La liste triée est: \n",tab2)
t2=time()
a2=t2-t1
print("le temps d'éxécution de tri à bulle est: \n",a2)
```

## 3) Algorithme par insertion:

*Implémentation du tri par insertion*

```python
from random import *
from time import *
def liste_alea(n):
    s=[0]*n
    for i in range(n):
        s[i]=randrange(1,100)
    return s
def tri_insertion(liste): 
    # Parcour de 1 à la taille du tab
    for i in range(1, len(liste)): 
        k = liste[i] 
        j = i-1
        while j >= 0 and k < liste[j] : 
                liste[j + 1] = liste[j] 
                j -= 1
        liste[j + 1] = k
```

*Programme principale pour tester le code ci-dessus et afficher le temps d'éxécution*

```python
tab3 = liste_alea(50)
t1=time()
print ("La liste initiale est: \n",tab3)
tri_insertion(tab3)
print ("La liste triée est: \n",tab3)
t2=time()
a3=t2-t1
print("le temps d'éxécution de tri par insertion est: \n",a3)

    
```

## 4) Tri par fusion :

*Implémentation du Tri par fusion*

```python
from random import *
from time import *
def liste_alea(n):
    s=[0]*n
    for i in range(n):
        s[i]=randrange(1,100)
    return s
def fusion(L1,L2):
    n1 = len(L1)
    n2 = len(L2)
    L12 = [0]*(n1+n2)
    i1 = 0
    i2 = 0
    i = 0
    while i1<n1 and i2<n2:
        if L1[i1] < L2[i2]:
            L12[i] = L1[i1]
            i1 += 1
        else:
            L12[i] = L2[i2]
            i2 += 1
        i += 1
    if i1<n1:
        while i1<n1:
            L12[i] = L1[i1]
            i1 += 1
            i += 1
    else:
        while i2<n2:
            L12[i] = L2[i2]
            i2 += 1
            i += 1
    return L12
def tri_fusion_recursif(liste):
    if len(liste) > 1:
        p = int(len(liste)/2)
        L1 = liste[0:p]
        L2 = liste[p:len(liste)]
        print("séparation : ",L1,L2)
        tri_fusion_recursif(L1)
        tri_fusion_recursif(L2)
        liste[:] = fusion(L1,L2)
        print("fusion : ",liste)
    
def tri_fusion(liste):
    M = list(liste)
    tri_fusion_recursif(M)
    return M
```

*Programme principale pour tester le code ci-dessus et afficher le temps d'éxécution*

```python
tab4 = liste_alea(50)
t1=time()
print ("La liste initiale est: \n",tab4)
tri_fusion(tab4)
t2=time()
a4=t2-t1
print("le temps d'éxécution de tri par fusion est: \n",a4)

    
```

## 5) Tri rapide :



*Implémentation du Tri rapide (Tri par partitionnement)*

```python
from random import *
from time import *
def liste_alea(n):
    s=[0]*n
    for i in range(n):
        s[i]=randrange(1,100)
    return s
def partition(liste,debut,fin):
    pivot = liste[fin]
    i = debut
    j = debut
    while j < fin:
        if liste[j] <= pivot:
            liste[i],liste[j] = liste[j],liste[i]
            i += 1
        j += 1
    liste[fin],liste[i] = liste[i],liste[fin]
    return i
def tri_partition_recursif(liste,debut,fin):
    if debut < fin:
        i = partition(liste,debut,fin)
        tri_partition_recursif(liste,debut,i-1)
        tri_partition_recursif(liste,i+1,fin)
def tri_partition(liste):
    tri_partition_recursif(liste,0,len(liste)-1)
    return liste
```

*Programme principale pour tester le code ci-dessus et afficher le temps d'éxécution*

```python
tab5 = liste_alea(20)
t1=time()
print ("La liste initiale est: \n",tab5)
tri_partition(tab5)
print ("La liste triée est: \n",tab5)
t2=time()
a5=t2-t1
print("le temps d'éxécution de tri rapide est: \n",a5)
```
